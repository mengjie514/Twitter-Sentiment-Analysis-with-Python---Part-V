# Project 5.0 Data Preparaton - Extracting Extended Tweets and Replace with Text

```python
import os
import json
import numpy as np
import pandas as pd
import numpy as np
os.chdir('\\\\file\\UsersM$\\mwa150\\Home\\data')
pd.set_option('display.max_colwidth', -1)
```

```python
# real_madrid datasets (pre-event)
data = []
with open("madrid_pre.json") as f:
    # use 'marid_post_week3 as a test sample'
    for line in f:
        try:
            tweet = json.loads(line)
            if tweet['text'].find('RT @')!=-1:
            # drop Retweeted tweet
                continue
            data.append(tweet)
        except:
            continue

with open('nmadrid_pre.json','w') as fp:
    json.dump(data,fp)

tweets_data_path=('nmadrid_pre.json')
tweets_data = []
tweets_file = open(tweets_data_path,'r')
tweets_data = json.load(tweets_file)
type(tweets_data)

# convert the json file to dataframe
df_pre = pd.DataFrame(tweets_data)

# extract extended_tweet and save into a new dataframe
ndf_pre = df_pre.dropna(subset=['extended_tweet'])
add_txt = []

for index, row in ndf_pre.iterrows():
    tx = (row['extended_tweet']['full_text'])   
    add_txt.append(tx)
    
ndf_pre['txt'] = add_txt

# dataframe cleaning
col_to_keep = ['created_at','text']
df_pre = df_pre[col_to_keep]
col_to_keep2 = ['created_at','text','txt']
ndf_pre = ndf_pre[col_to_keep2]

df1 = df_pre.merge(ndf_pre,how='left')
df1.head()

df1.txt.fillna(df1.text, inplace=True)
del df1['text']
```
